```python
import pandas as pd
star_wars = pd.read_csv("StarWars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
star_wars.head(10)
```

```python
yes_no = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

star_wars.head(10)
```

```python
star_wars.columns[3:9]
```

```python
import numpy as np
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
    
    
star_wars.head(10)
```

```python

star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]].head(10)    
```

```python
x = star_wars[star_wars.columns[9:15]].mean()
print(x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

ind = np.arange(6)
width = 0.35  

fig,ax = plt.subplots()
ax.set_xticklabels(star_wars.columns[9:15])
ax.set_xticks(ind + width / 2)

ax.bar(ind, star_wars[star_wars.columns[9:15]].mean())

plt.show()
```

```python
male = star_wars[star_wars['Gender'] == 'Male']
female = star_wars[star_wars['Gender'] == 'Female']

N = 6

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

male_mean = male[male.columns[9:15]].mean()
female_mean = female[female.columns[9:15]].mean()

fig, ax = plt.subplots()
malebar = ax.bar(ind, male_mean, width, color='r')

femalebar = ax.bar(ind + width, female_mean, width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Ranking the Star Wars Episode by gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(male[male.columns[9:15]])

plt.show()
```

```python
male = star_wars[star_wars['Gender'] == 'Male']
female = star_wars[star_wars['Gender'] == 'Female']

N = 6

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

male_sum = male[male.columns[3:9]].sum()
female_sum = female[female.columns[3:9]].sum()

fig, ax = plt.subplots()
malebar = ax.bar(ind, male_sum, width, color='r')

femalebar = ax.bar(ind + width, female_sum, width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Seen the Star Wars Episode by gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(male[male.columns[3:9]])

plt.show()
```
