```python
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
plt.plot(women_degrees["Year"],women_degrees["Biology"])

plt.show()
```

```python
fig =  plt.figure()
ax = fig.add_subplot(111)

ax.plot(women_degrees["Year"],women_degrees["Biology"],label="Women",color="blue")
ax.plot(women_degrees["Year"],100 - women_degrees["Biology"],label="Men",color="green")
ax.set_title("Percentage of Biology Degrees Awarded By Gender")
plt.legend(loc="upper right")
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(16, 20))

for sp in range(0,6):
    ax = fig.add_subplot(6,3,(sp*3)+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #reduce y-label shows only 0 and 100
    ax.set_yticks([0,100])
    #generate a horizontal line across
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.tick_params(labelbottom='on')
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')

for lp in range(0,5):
    ax = fig.add_subplot(6,3,(lp*3)+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[lp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[lp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #reduce y-label shows only 0 and 100
    ax.set_yticks([0,100])
    #generate a horizontal line across
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(lib_arts_cats[lp])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    
    if lp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')  
    elif lp == 4:
        ax.tick_params(labelbottom='on')
        
for op in range(0,6):
    ax = fig.add_subplot(6,3,(op*3)+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[op]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[op]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #reduce y-label shows only 0 and 100
    ax.set_yticks([0,100])
    #generate a horizontal line across
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(other_cats[op])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    
    if op == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif op == 5:
        ax.tick_params(labelbottom='on')
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')     

#Save all charts to image file        
plt.savefig('gender_degrees.png')        
plt.show()
```
