```python
import sqlite3
import pandas as pd
conn = sqlite3.connect("factbook.db")

q = "SELECT * FROM sqlite_master WHERE type='table';"
pd.read_sql_query(q, conn)
```

```python
query = "SELECT * FROM facts LIMIT 5"
pd.read_sql_query(query, conn)
```

```python
q = "select min(population), max(population), min(population_growth), max(population_growth) from facts"
pd.read_sql_query(q,conn)
```

```python
q = "select * from facts where population = 0"
pd.read_sql_query(q,conn)
```

```python
q = "select * from facts where population = 7256490011"
pd.read_sql_query(q,conn)
```

```python
q = "select population, population_growth,birth_rate, death_rate from facts"

result  = pd.read_sql_query(q,conn)

import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(2,2,1)
ax.set_title("population")
result["population"].hist(ax=ax)

ax = fig.add_subplot(2,2,2)
ax.set_title("population_growth")
result["population_growth"].hist(ax=ax)

ax = fig.add_subplot(2,2,3)
ax.set_title("birth_rate")
result["birth_rate"].hist(ax=ax)

ax = fig.add_subplot(2,2,4)
ax.set_title("death_rate")
result["death_rate"].hist(ax=ax)

fig.tight_layout() # Adjjust layout automatically
plt.show()
```
